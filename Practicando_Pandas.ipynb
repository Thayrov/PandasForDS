```python
#Supongamos que estamos probando las monedas que utilizarán los árbitros de fútbol en las competiciones de la Copa del Mundo. Actualmente, estas monedas están personalizadas y se utilizan en determinados momentos del juego para decidir, por suerte, una disputa. Nuestro objetivo es verificar que las monedas que se utilizarán en los juegos no sean viciadas.

#Para ello se realizaron pruebas con cinco monedas y los resultados fueron los siguientes:
```

```python
m1 = 'CCcCCccCCCccCcCccCcCcCCCcCCcccCCcCcCcCcccCCcCcccCc'
m2 = 'CCCCCccCccCcCCCCccCccccCccCccCCcCccCcCcCCcCccCccCc'
m3 = 'CccCCccCcCCCCCCCCCCcccCccCCCCCCccCCCcccCCCcCCcccCC'
m4 = 'cCCccCCccCCccCCccccCcCcCcCcCcCcCCCCccccCCCcCCcCCCC'
m5 = 'CCCcCcCcCcCCCcCCcCcCCccCcCCcccCccCCcCcCcCcCcccccCc'
```

```python
#Arriba, tenemos el resultado de 50 lanzamientos de cada moneda (m1, m2, m3, m4 y m5), donde c representa la ocurrencia del evento CARA y C representa la ocurrencia del evento CROWN (sello o corona).

#Para sacar nuestras conclusiones, necesitamos construir un DataFrame, Donde Frec. C y Frec. c son, respectivamente, las frecuencias de CROWN y CARA de cada prueba.
```

##Identificando y Excluyendo Outliers

```python
datos = pd.read_csv('alquiler_residencial.csv', sep = ';')
datos.head(10)
```

```python
datos.boxplot(['Valor'])
```

```python
datos[datos['Valor'] >= 500000]
```

```python
valor = datos['Valor']
```

```python
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 -1.5 * IIQ
limite_superior = Q3 +1.5 * IIQ
```

```python
seleccion = ((valor >= limite_inferior) & (valor <= limite_superior))
```

```python
datos_new = datos[seleccion]
```

```python
datos_new.boxplot(['Valor'])
```

```python
datos.hist(['Valor'])
```

```python
datos_new.hist(['Valor'])
```

##Identificando y Excluyendo Outliers por Grupo

```python
datos.boxplot(['Valor'], by = ['Tipo'])
```

```python
datos.head(10)
```

```python
grupo_tipo = datos.groupby('Tipo')['Valor']
```

```python
grupo_tipo.groups
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 -1.5 * IIQ
limite_superior = Q3 +1.5 * IIQ
```

```python
limite_inferior
```

```python
limite_superior
```

```python
datos_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  eh_tipo = datos['Tipo'] == tipo
  eh_dentro_limite = (datos['Valor'] >= limite_inferior[tipo]) & (datos['Valor'] <= limite_superior[tipo])
  seleccion = eh_tipo & eh_dentro_limite
  datos_seleccion = datos[seleccion]
  datos_new = pd.concat([datos_new, datos_seleccion])
```

```python
datos_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
datos_new.to_csv('alquiler_residencial_sin_outliers.csv', sep=';', index =False)
```
